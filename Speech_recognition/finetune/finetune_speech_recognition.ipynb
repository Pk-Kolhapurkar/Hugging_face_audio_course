```python
from transformers import WhisperProcessor

processor = WhisperProcessor.from_pretrained(
    "openai/whisper-small", language="sinhalese", task="transcribe"
)
```

```python
from datasets import Audio

sampling_rate = processor.feature_extractor.sampling_rate   #retrieves the sampling rate expected by the Whisper model.
common_voice = common_voice.cast_column("audio", Audio(sampling_rate=sampling_rate))  #Convert the Audio Column to the Required Sampling Rate
```

```python
def prepare_dataset(example):
    audio = example["audio"]  #Extract Audio Data
    #The processor converts raw audio into features needed for the model.This step tokenizes the text and extracts audio features that the model will use for training.
    example = processor(
        audio=audio["array"],
        sampling_rate=audio["sampling_rate"],
        text=example["sentence"],
    )

    # compute input length of audio sample in seconds
    example["input_length"] = len(audio["array"]) / audio["sampling_rate"]

    return example
```

```python
#This line applies the prepare_dataset function to every sample in the common_voice dataset
common_voice = common_voice.map(
    prepare_dataset, remove_columns=common_voice.column_names["train"], num_proc=1
)
```

```python
max_input_length = 30.0


def is_audio_in_length_range(length):
    return length < max_input_length
```

```python
common_voice["train"] = common_voice["train"].filter(
    is_audio_in_length_range,
    input_columns=["input_length"],
)
```

This class is a custom data collator designed for automatic speech recognition (ASR) models that follow a sequence-to-sequence (Seq2Seq) approach. It ensures that both audio inputs and tokenized text labels are properly padded before being fed into a model.

```python
import torch

from dataclasses import dataclass
from typing import Any, Dict, List, Union


@dataclass
class DataCollatorSpeechSeq2SeqWithPadding:
    processor: Any    #This stores the processor (which includes both feature_extractor for audio and tokenizer for text).


    def __call__(     #This method processes a batch of examples (audio inputs + text labels) and returns them as padded tensors.
        self, features: List[Dict[str, Union[List[int], torch.Tensor]]]
    ) -> Dict[str, torch.Tensor]:
        # split inputs and labels since they have to be of different lengths and need different padding methods
        # first treat the audio inputs by simply returning torch tensors

        '''
    Extracts input_features (i.e., the processed audio features) from each example.
    Uses the feature extractor to pad the audio inputs so that all have the same length.
    Converts them into PyTorch tensors (return_tensors="pt").'''
        input_features = [
            {"input_features": feature["input_features"][0]} for feature in features
        ]
        batch = self.processor.feature_extractor.pad(input_features, return_tensors="pt")

      '''
      Extracts the labels (i.e., tokenized text sequences) from each example.
      Uses the tokenizer to pad the labels to the same length.
      Converts them into PyTorch tensors.'''
        # get the tokenized label sequences
        label_features = [{"input_ids": feature["labels"]} for feature in features]
        # pad the labels to max length
        labels_batch = self.processor.tokenizer.pad(label_features, return_tensors="pt")
      '''
      The attention_mask tells which tokens are real words (1) and which are padding (0).
masked_fill(labels_batch.attention_mask.ne(1), -100):
Replaces all padding tokens (0s) with -100, which is ignored during loss calculation in most transformer models.
'''
        # replace padding with -100 to ignore loss correctly
        labels = labels_batch["input_ids"].masked_fill(
            labels_batch.attention_mask.ne(1), -100
        )
      '''
      Some tokenizers add a BOS token (<s>) at the start of sequences.
If all labels start with BOS, it is removed since it will be re-added later.

'''
        # if bos token is appended in previous tokenization step,
        # cut bos token here as it's append later anyways
        if (labels[:, 0] == self.processor.tokenizer.bos_token_id).all().cpu().item():
            labels = labels[:, 1:]
      '''
      The processed labels are added back to the batch dictionary.
The final output batch includes padded audio features and padded text labels, ready for model training.
'''
        batch["labels"] = labels

        return batch
```

```python
data_collator = DataCollatorSpeechSeq2SeqWithPadding(processor=processor)
```

```python
import evaluate

metric = evaluate.load("wer")
```

```python
from transformers.models.whisper.english_normalizer import BasicTextNormalizer

normalizer = BasicTextNormalizer()#BasicTextNormalizer is used to normalize text (e.g., removing punctuation, converting to lowercase).This helps in computing a normalized WER, which ignores minor variations in text.

'''
It calculates two versions of WER:
Orthographic WER (wer_ortho): Direct comparison of predicted vs. reference text.
Normalized WER (wer): Uses BasicTextNormalizer to clean the text before comparison.
'''
def compute_metrics(pred):
    pred_ids = pred.predictions #pred_ids: Model's predicted token IDs.
    label_ids = pred.label_ids #label_ids: Ground truth token IDs.

    # replace -100 with the pad_token_id
    label_ids[label_ids == -100] = processor.tokenizer.pad_token_id  #The training process ignores tokens with -100, so we replace them with the padding token ID in evaluation.

    # we do not want to group tokens when computing the metrics.Converts predicted and reference token IDs into human-readable text.skip_special_tokens=True removes special tokens like <s> (start) and </s> (end).

    pred_str = processor.batch_decode(pred_ids, skip_special_tokens=True)
    label_str = processor.batch_decode(label_ids, skip_special_tokens=True)

    # compute orthographic wer
    wer_ortho = 100 * metric.compute(predictions=pred_str, references=label_str)

    # compute normalised WER
    pred_str_norm = [normalizer(pred) for pred in pred_str]
    label_str_norm = [normalizer(label) for label in label_str]#Helps remove minor formatting differences (e.g., "Hello!" → "hello").

    # filtering step to only evaluate the samples that correspond to non-zero references.Removes samples where the reference text is empty (to prevent errors in WER calculation).

    pred_str_norm = [
        pred_str_norm[i] for i in range(len(pred_str_norm)) if len(label_str_norm[i]) > 0
    ]
    label_str_norm = [
        label_str_norm[i]
        for i in range(len(label_str_norm))
        if len(label_str_norm[i]) > 0
    ]

    wer = 100 * metric.compute(predictions=pred_str_norm, references=label_str_norm)

    return {"wer_ortho": wer_ortho, "wer": wer}
```

```python
from transformers import WhisperForConditionalGeneration

model = WhisperForConditionalGeneration.from_pretrained("openai/whisper-small")
```

This code modifies the Whisper model's settings to improve training efficiency and customizes text generation for Sinhala language transcription.



```python
from functools import partial #partial() is used to pre-configure a function with default parameters.

# disable cache during training since it's incompatible with gradient checkpointing
model.config.use_cache = False

# set language and task for generation and re-enable cache
model.generate = partial(
    model.generate, language="sinhalese", task="transcribe", use_cache=True
)
```

```python
from transformers import Seq2SeqTrainingArguments

training_args = Seq2SeqTrainingArguments(
    output_dir="./whisper-small-dv",  # name on the HF Hub
    per_device_train_batch_size=16,
    gradient_accumulation_steps=1,  # increase by 2x for every 2x decrease in batch size.If you reduce batch size, increase this to maintain stability
    learning_rate=1e-5,
    lr_scheduler_type="constant_with_warmup",
    warmup_steps=50,  #The first 50 steps gradually increase LR to avoid sudden changes.
    max_steps=500,  # increase to 4000 if you have your own GPU or a Colab paid plan
    gradient_checkpointing=True,  ##Reduces memory usage by not storing all intermediate activations.
    fp16=True,  #Uses 16-bit floating point precision for faster training and lower memory usage.
    fp16_full_eval=True,
    evaluation_strategy="steps",
    per_device_eval_batch_size=16,
    predict_with_generate=True,  #Uses Whisper's generate() function for predictions.
    generation_max_length=225,
    save_steps=500,  #Saves the model every 500 steps.
    eval_steps=500,  #Runs evaluation every 500 steps.
    logging_steps=25,
    report_to=["tensorboard"],  #Sends logs to TensorBoard for visualization.
    load_best_model_at_end=True,
    metric_for_best_model="wer",
    greater_is_better=False,  #Lower WER is better, so we minimize it.
    push_to_hub=True,
)
```

```python
from transformers import Seq2SeqTrainer

trainer = Seq2SeqTrainer(
    args=training_args,
    model=model,
    train_dataset=common_voice["train"],
    eval_dataset=common_voice["test"],
    data_collator=data_collator,
    compute_metrics=compute_metrics,
    tokenizer=processor,
)
```

```python
trainer.train()
```

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("automatic-speech-recognition", model="Prathamesh1420/whisper-small-dv")
```

```python
'''def transcribe_speech(filepath):
    output = pipe(
        filepath,
        max_new_tokens=256,
        generate_kwargs={
            "task": "transcribe",
            "language": "sinhalese",
        },  # update with the language you've fine-tuned on
        chunk_length_s=30,
        batch_size=8,
    )
    return output["text"]'''
import os

def transcribe_speech(filepath):
    if not os.path.exists(filepath):
        raise ValueError(f"File not found: {filepath}")  # Debugging check

    output = pipe(
        filepath,
        generate_kwargs={
            "task": "transcribe",
            "language": "sinhalese",
            "max_new_tokens": 256,
        },
        chunk_length_s=30,
        batch_size=8,
    )
    return output["text"]
```

```python
import gradio as gr

demo = gr.Blocks()

mic_transcribe = gr.Interface(
    fn=transcribe_speech,
    inputs=gr.Audio(sources="microphone", type="filepath"),
    outputs=gr.components.Textbox(),
)

file_transcribe = gr.Interface(
    fn=transcribe_speech,
    inputs=gr.Audio(sources="upload", type="filepath"),
    outputs=gr.components.Textbox(),
)
```

```python
with demo:
    gr.TabbedInterface(
        [mic_transcribe, file_transcribe],
        ["Transcribe Microphone", "Transcribe Audio File"],
    )

demo.launch(debug=True)
```
